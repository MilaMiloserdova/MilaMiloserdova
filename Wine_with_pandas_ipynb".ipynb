```python
sugar_filter = bellevin_df['sugar'] == "Сухое"
sugar_filter1 = bellevin_df['sugar'] == "Брют"
```

```python
sugar_filter2 = bellevin_df['sugar'] == "Экстра Брют"
```

```python
sugar_filter3 = bellevin_df['sugar'] == "Полусладкое"
```

```python
sugar_filter4 = bellevin_df['sugar'] == "Полусухое"
```

```python
regions_sugar = bellevin_df[sugar_filter+sugar_filter1+sugar_filter2+sugar_filter3+sugar_filter4]
```

```python
regions_sugar.value_counts()
```

```python
regions_sugar['sugar'].value_counts()
```

```python
regions_sugar['sugar'].value_counts().plot(kind='pie', figsize=(10,10), title='Sugar Level:\nMost Frequent winemaking style in the world')
```
